```
from snowflake.snowpark.context import get_active_session
import pandas as pd
session = get_active_session()
df = session.sql("SELECT * FROM KAGGLE_LOGISTICS.DEV_STAGING.STG_RAW__GLOBAL_SUPPLY_CHAIN_RISK_2024_2026").to_pandas()
df.head()
```

```
df_Rail = df[df['TRANSPORT_MODE']=='Rail']
df_Air = df[df['TRANSPORT_MODE']=='Air']
df_Sea = df[df['TRANSPORT_MODE']=='Sea']
df_Road = df[df['TRANSPORT_MODE']=='Road']
df_list = [df_Rail, df_Air, df_Sea, df_Road]
for i in df_list:
    print(i.shape)
    display(i.describe())
```

```
for i in df_list:
    display(i.head())
```

```
import seaborn as sns
from matplotlib import pyplot as plt
for i in df_list:
    i["DESTINATIONS"] = i["ORIGIN_PORT"] + "~" + i["DESTINATION_PORT"]
    i.groupby("DESTINATIONS")[["WEIGHT_MT"]].agg({"WEIGHT_MT": "mean"}).reset_index().sort_values(by = "WEIGHT_MT")
    sns.barplot(
        data = i, 
        x = "DESTINATIONS",
        y = "WEIGHT_MT"
    )
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()
    
```
